Part3 データ整理

```python
#予測に必要のない列を削除＆節数とキックオフ時刻点数差分を編集
data["節数"]=data["節数"].str.replace('第', '')
data["K/O時刻"]=data["K/O時刻"].str[:2]
score=data.スコア.str.split('-', expand=True)
score=score[[0,1]].astype(np.int64)
data["点数(ホームーアウェイ)"]=score[0]-score[1]
data=data.drop(["年度","節","試合日","ホーム","インターネット中継・TV放送","年/月/日","アウェイ","スコア"],axis=1)
data.head()
```

```python
#入場者数の平均を出力
visitors=int(data.入場者数.mean())
print(visitors)
#各試合の入場者数が平均より多ければlarge、少なければsmallと表示する列を作成
data.loc[data['入場者数'] > visitors, 'large or small'] = "large"
data.loc[data['入場者数'] < visitors, 'large or small'] = "small"
large_number=data=="large"
small_number=data=="small"
print(large_number.values.sum())
print(small_number.values.sum())
data.head()
```

```python
#卒研＿入場者数.ipynbから各地域のキャパを取得
#入場者数は入場者数＝入場者数/キャパ
capa=pd.read_csv("capa.csv").drop(["Unnamed: 0"],axis=1)
capacity=capa[capa.クラブ名==team_name]
capacity=capacity.iat[0,1]#収容人数
data["入場者数"]=data["入場者数"]/int(capacity)
print(int(capacity))
data
```

```python
#天気データを平均差で表す
#降水量と日照時間はあるかなしなので、偏差をとる必要はない
temperature＿mean=int(data["平均気温(℃)"].mean())
print(temperature＿mean)
data["平均気温(℃)"]=data["平均気温(℃)"]-temperature＿mean
data.head()
```

```python
#largeとsmallそれぞれの個数を計測
print(data["large or small"].value_counts())
#決定木を用いて分類
clf = tree.DecisionTreeClassifier(max_depth = 3)
y=data['large or small'].values
x=data.drop(["入場者数","large or small"], axis=1)
x.head()
```

```python
clf=clf.fit(x,y)
predicted = clf.predict(x)
print('=============== predicted ===============')
print(predicted)
print('============== correct_ans ==============')
print(y)
print('=============== id_rate =================')
print(sum(predicted == y) / len(x))
```
